# 凸包について

---

今回は凸包についてコードも含めながら解説していきます．

凸包は最適化問題などや，統計的学習理論などで出てくるフレーズです．

まず，数式を書いていきます．

まず，点の集合を$P$とします．

その集合に対しての凸包(conv)とは

$$
\operatorname{conv}(P)=\left\{\sum_{i=1}^n \lambda_i p_i \mid \lambda_i \geq 0, \sum_{i=1}^n \lambda_i=1, p_i \in P\right\}
$$

この式の直感的な理解としては，同じ集合内の点が複数あり，その点にある重みをつけて，重みをつけた点たちを全て囲むような最小な凸多角形のことをいいます．

これを考えると何がいいかというと，線形問題を例に出すと，凸包に対するある目的関数があるとすると，最適解は凸包上の境界にあることが知られています．

しかしながら，疑問としては，この多角形の形は重みによりますね．

凸包の求め方は，色々なアルゴリズムはあることが知られています．

---

ここから二次元の時，三次元の凸包をコードで書いていきます．

```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt

points = np.array([[0, 3], [0, 1], [-4, 1], [5, 0.5], [2, 0.5]])

# 凸包を計算する
hull = ConvexHull(points)

# 凸包の頂点を取得する
vertices = hull.vertices


plt.plot(points[:,0], points[:,1], 'o')
for sim in hull.simplices:
    plt.plot(points[sim, 0], points[sim, 1], 'k-')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

points = np.array([
    [0, 0, 0],
    [0, 5, 0],
    [1, 0, 0],
    [0, 2, 1],
    [1, 1, 3],
    [1, 1, 1],
    [1, 2, 1],
    [1, 1, 3]
])

# 凸包を計算する
hull = ConvexHull(points)

# 凸包の頂点の座標を取得する
vertices = hull.points[hull.vertices]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(points[:,0], points[:,1], points[:,2], triangles=hull.simplices, color='gray', alpha=0.2)
ax.scatter(vertices[:,0], vertices[:,1], vertices[:,2], color='red', s=50)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

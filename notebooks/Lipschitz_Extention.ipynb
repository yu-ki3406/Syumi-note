# リプシッツ連続の拡張

---

今回はこの[論文](https://openreview.net/pdf?id=BJgnXpVYwS)で書かれているリプシッツ連続の拡張版を説明していきます。

リプシッツ連続の簡単な説明は[こちら](https://github.com/yu-ki3406/Syumi-note/blob/main/notebooks/Lipschitz_continuity.ipynb)に書いてあります。

まず、定義から確認していきましょう。

微分可能な実関数$f$がL-smoothとは次のことを意味します。

$$
\|\nabla f(x)-\nabla f(y)\| \leq L\|x-y\|, \quad \text { for all } x, y \in \mathbb{R}^d \text {. }
$$

$$
\left\|\nabla^2 f(x)\right\| \leq L, \forall x \in \mathbb{R}^d
$$

しかしながら、上の定義を確認すれば明らかで、もし、局所的で大きな傾きがあった場合、Lは大きくとらなければいけません。

大きいLをとることはアルゴリズムの最適化を遅くします。

---

そこで、提案されたのが、($L_0,L_1$)-smoothです。

$$
\left\|\nabla^2 f(x)\right\| \leq L_0+L_1\|\nabla f(x)\|
$$

$L_0,L_1$は定数で、$L_0 << L \text{かつ} L_1 << L$です。

これが意味していることは、勾配が大きいところでは、大きい値で抑え、小さい値では小さく抑えられるということです。

---

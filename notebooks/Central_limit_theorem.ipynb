# 中心極限定理(Central Limit Theorem)を説明します．


##### まず式の説明からしていきます．



$$ \Large \bar{X} \sim N(\mu, \frac{\sigma^2}{n}) $$

ここで, $\bar{X}$は $n$ 個の独立な同分布な母集団から抽出した標本平均, $\mu$は母集団の平均, $\sigma^2$は母集団の分散,$N(\mu, \frac{\sigma^2}{n})$は平均 $\mu$,分散 $\frac{\sigma^2}{n}$ の正規分布を表します.

中心極限定理とは, **標本の大きさが十分大きい**場合, **標本平均 $\bar{X}$ は元の母集団の分布の平均**に近づくことを意味します.

ここで標本の大きさによってどのくらい標本分布の形が異なるか見ていきます.

```python
import numpy as np
import matplotlib.pyplot as plt

#ガウス分布[0,1]の分布を想定
def generate_distribution(size):
    return np.random.rand(size)

#標本平均を出す関数
def calculate_sample_mean(sample_size, number_of_samples):
    sample_means = []
    for _ in range(number_of_samples):
        sample = generate_distribution(sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)
    return sample_means

#今回のサンプル数は次のように変化させた
sample_sizes = [1, 5, 10, 50, 100]
number_of_samples = 1000

for sample_size in sample_sizes:
    sample_means = calculate_sample_mean(sample_size, number_of_samples)
    plt.hist(sample_means,alpha=0.7)
    plt.title(f'sample size: {sample_size}')
    plt.xlim(0.0,1.0)
    plt.xlabel('random variable')
    plt.ylabel('number of observations')
    plt.show()
```

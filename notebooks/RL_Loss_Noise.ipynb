# 論文を解説

今回はこちらの[論文(2011 Amir)](https://link.springer.com/article/10.1007/s10994-011-5254-7)に書いてあった強化学習を行う際に損失関数にある一定の定数を含める理由を説明していきます．

コードは書かずに自分用の備忘録としてまとめていきます．

---

強化学習でいいモデルを作るには次の問題を考えます，

$Q_k$ は真の価値関数とします．

$$
\underset{1 \leq k \leq P}{\operatorname{argmin}}\left\|Q_k-T^*Q_k\right\|_\nu^2 .
$$

しかしながら最適ベルマンオペレーターを使えない時が学習の設定上あるので次の問題を考えます．

$Q_k$ は真の価値関数， $\tilde{Q}_k$ は真の価値関数に最適ベルマンオペレーターを採用したもの ($T^*Q_k$) を推定したもの，kは推定価値関数の集合($\tilde{Q}_{1,2,3...,k}$)から選んだindexとします．



$$
\underset{1 \leq k \leq P}{\operatorname{argmin}}\left\|Q_k-\tilde{Q}_k\right\|_n^2 .
$$

下に上で出てきた関数の関係性を示します．

しかしながら上の問題を解くことを考えると次の画像のような問題が起きてしまいます．

$\tilda Q^{(1)}_k$ と $\tilda Q^{(2)}_k$ は $T^*Q_k$ を推定した際の予測値です．

もし上の問題を解いてしまうと $\tilde Q^{(2)}_k$ が $T^*Q_k$ を推定する関数の中で良いものとして捉えられてしまいますね，

しかしながら実際は $\tilde Q^{(1)}_k$ が $\tilde Q^{(2)}_k$ よりも良い関数ですね，

このことから三角不等式を考えて次の問題に書き換えます．

$$
\left\|Q_k-T^* Q_k\right\|_\nu^2 \leq 2\left[\left\|Q_k-\tilde{Q}_k\right\|_\nu^2+\left\|T^* Q_k-\tilde{Q}_k\right\|_\nu^2\right]
$$

この上の式の $|T^* Q_k-\tilde{Q}_k|_\nu^2$ を何らかの関数若しくは定数で押さえつけるという作業を行えば，この問題は解けるというようになっていきます．
